```python
import cv2
from ultralytics import YOLO

# Load the trained model
model = YOLO(r"C:\Users\tahar\runs\detect\train14\weights\best.pt")  

# Initialize webcam (0 for default webcam)
cap = cv2.VideoCapture(0)

while True:
    # Read a frame from the webcam
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame")
        break
    
    # Perform inference on the frame
    results = model(frame)

    # Render the results on the frame (bounding boxes, labels, etc.) using plot()
    frame = results[0].plot()  # Access the first result and render it

    # Show the frame with predictions (bounding boxes)
    cv2.imshow("Real-Time Detection", frame)

    # Exit the loop when the 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
